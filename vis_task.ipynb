```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from maml.datasets.task import plot_task
```

```python
from maml.datasets.miniimagenet import MiniimagenetMetaDataset
dataset = MiniimagenetMetaDataset(
    root='data',
    img_side_len=84,
    num_classes_per_batch=5,
    num_samples_per_class=5,
    num_total_batches=5,
    num_val_samples=5,
    meta_batch_size=2,
    train=True,
    num_workers=4,
    device='cuda')
```

```python
for train_batch, val_batch in iter(dataset):
    print('a')
train_task, val_task = train_batch[0], val_batch[0]
plot_task(train_task)
plot_task(val_task)
```
